上手くドライバが入っていたのか、特に苦労もしなかった。!を前に付けることで、UbuntuのコマンドをJupyter内で実行できるらしい。
!fswebcam  --no-banner --save {orig_img_path} -d /dev/video0 2> /dev/null
はwebカメラからのデータをorig_img_pathに記録するコマンドらしい。

今回はFPGAとなんの関係もない。

```python
from PIL import Image as Image
import matplotlib.pyplot as plt
import numpy as np

orig_img_path = "/home/xilinx/jupyter_notebooks/Examples/data/webcam.jpg"
!fswebcam  --no-banner --save {orig_img_path} -d /dev/video0 2> /dev/null

#画像の読み込み
im = Image.open(orig_img_path)

#画像をarrayに変換
im_list = np.asarray(im)
#貼り付け
plt.imshow(im_list)
#表示
plt.show()
```
